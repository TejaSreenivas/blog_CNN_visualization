```
########################################################################################
# Davi Frossard, 2016                                                                  #
# VGG16 implementation in TensorFlow                                                   #
# Details:                                                                             #
# http://www.cs.toronto.edu/~frossard/post/vgg16/                                      #
#                                                                                      #
# Model from https://gist.github.com/ksimonyan/211839e770f7b538e2d8#file-readme-md     #
# Weights from Caffe converted using https://github.com/ethereon/caffe-tensorflow      #
########################################################################################

import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
#from scipy.misc import imread, imresize
#from imagenet_classes import class_names


class vgg16:
    def __init__(self, imgs, wts=None, sess=None):
        self.imgs = imgs
        self.trainable = False
        if wts is not None and sess is not None:
            self.weights = np.load(wts)
            #self.load_weights(wts, sess)
        self.convlayers()
        self.fc_layers()
        self.probs = tf.nn.softmax(self.fc3l)
        if sess is not None:
            self.model_graph = sess.graph


    def convlayers(self):
        self.parameters = []

        # zero-mean input
        with tf.variable_scope('preprocess',reuse=tf.AUTO_REUSE) as scope:
            mean = tf.constant([123.68, 116.779, 103.939], dtype=tf.float32, shape=[1, 1, 1, 3], name='img_mean')
            images = self.imgs-mean

        # conv1_1
        with tf.variable_scope('conv1_1',reuse=tf.AUTO_REUSE) as scope:
            kernel = tf.get_variable(initializer = self.weights['conv1_1_W'], name = 'weights', trainable= self.trainable)
            conv = tf.nn.conv2d(images, kernel, [1, 1, 1, 1], padding='SAME')
            biases = tf.get_variable(initializer = self.weights['conv1_1_b'], name='biases', trainable= self.trainable)
            out = tf.nn.bias_add(conv, biases)
            self.conv1_1 = tf.nn.relu(out)
            self.parameters += [kernel, biases]

        # conv1_2
        with tf.variable_scope('conv1_2',reuse=tf.AUTO_REUSE) as scope:
            kernel = tf.get_variable(initializer = self.weights['conv1_2_W'], name='weights', trainable= self.trainable)
            conv = tf.nn.conv2d(self.conv1_1, kernel, [1, 1, 1, 1], padding='SAME')
            biases = tf.get_variable(initializer = self.weights['conv1_2_b'], name='biases', trainable= self.trainable)
            out = tf.nn.bias_add(conv, biases)
            self.conv1_2 = tf.nn.relu(out)
            self.parameters += [kernel, biases]

        # pool1
        self.pool1 = tf.nn.max_pool(self.conv1_2,
                               ksize=[1, 2, 2, 1],
                               strides=[1, 2, 2, 1],
                               padding='SAME',
                               name='pool1')

        # conv2_1
        with tf.variable_scope('conv2_1',reuse=tf.AUTO_REUSE) as scope:
            kernel = tf.get_variable(initializer = self.weights['conv2_1_W'], name='weights', trainable= self.trainable)
            conv = tf.nn.conv2d(self.pool1, kernel, [1, 1, 1, 1], padding='SAME')
            biases = tf.get_variable(initializer = self.weights['conv2_1_b'], name='biases', trainable= self.trainable)
            out = tf.nn.bias_add(conv, biases)
            self.conv2_1 = tf.nn.relu(out)
            self.parameters += [kernel, biases]

        # conv2_2
        with tf.variable_scope('conv2_2',reuse=tf.AUTO_REUSE) as scope:
            kernel = tf.get_variable(initializer = self.weights['conv2_2_W'], name='weights', trainable= self.trainable)
            conv = tf.nn.conv2d(self.conv2_1, kernel, [1, 1, 1, 1], padding='SAME')
            biases = tf.get_variable(initializer = self.weights['conv2_2_b'], name='biases', trainable= self.trainable)
            out = tf.nn.bias_add(conv, biases)
            self.conv2_2 = tf.nn.relu(out)
            self.parameters += [kernel, biases]

        # pool2
        self.pool2 = tf.nn.max_pool(self.conv2_2,
                               ksize=[1, 2, 2, 1],
                               strides=[1, 2, 2, 1],
                               padding='SAME',
                               name='pool2')

        # conv3_1
        with tf.variable_scope('conv3_1',reuse=tf.AUTO_REUSE) as scope:
            kernel = tf.get_variable(initializer = self.weights['conv3_1_W'], name='weights', trainable= self.trainable)
            conv = tf.nn.conv2d(self.pool2, kernel, [1, 1, 1, 1], padding='SAME')
            biases = tf.get_variable(initializer = self.weights['conv3_1_b'], name='biases', trainable= self.trainable)
            out = tf.nn.bias_add(conv, biases)
            self.conv3_1 = tf.nn.relu(out)
            self.parameters += [kernel, biases]

        # conv3_2
        with tf.variable_scope('conv3_2',reuse=tf.AUTO_REUSE) as scope:
            kernel = tf.get_variable(initializer = self.weights['conv3_2_W'], name='weights', trainable= self.trainable)
            conv = tf.nn.conv2d(self.conv3_1, kernel, [1, 1, 1, 1], padding='SAME')
            biases = tf.get_variable(initializer = self.weights['conv3_2_b'], name='biases', trainable= self.trainable)
            out = tf.nn.bias_add(conv, biases)
            self.conv3_2 = tf.nn.relu(out)
            self.parameters += [kernel, biases]

        # conv3_3
        with tf.variable_scope('conv3_3',reuse=tf.AUTO_REUSE) as scope:
            kernel = tf.get_variable(initializer = self.weights['conv3_3_W'], name='weights', trainable= self.trainable)
            conv = tf.nn.conv2d(self.conv3_2, kernel, [1, 1, 1, 1], padding='SAME')
            biases = tf.get_variable(initializer = self.weights['conv3_3_b'], name='biases', trainable= self.trainable)
            out = tf.nn.bias_add(conv, biases)
            self.conv3_3 = tf.nn.relu(out)
            self.parameters += [kernel, biases]

        # pool3
        self.pool3 = tf.nn.max_pool(self.conv3_3,
                               ksize=[1, 2, 2, 1],
                               strides=[1, 2, 2, 1],
                               padding='SAME',
                               name='pool3')

        # conv4_1
        with tf.variable_scope('conv4_1',reuse=tf.AUTO_REUSE) as scope:
            kernel = tf.get_variable(initializer = self.weights['conv4_1_W'], name='weights', trainable= self.trainable)
            conv = tf.nn.conv2d(self.pool3, kernel, [1, 1, 1, 1], padding='SAME')
            biases = tf.get_variable(initializer = self.weights['conv4_1_b'], name='biases', trainable= self.trainable)
            out = tf.nn.bias_add(conv, biases)
            self.conv4_1 = tf.nn.relu(out)
            self.parameters += [kernel, biases]

        # conv4_2
        with tf.variable_scope('conv4_2',reuse=tf.AUTO_REUSE) as scope:
            kernel = tf.get_variable(initializer = self.weights['conv4_2_W'], name='weights', trainable= self.trainable)
            conv = tf.nn.conv2d(self.conv4_1, kernel, [1, 1, 1, 1], padding='SAME')
            biases = tf.get_variable(initializer = self.weights['conv4_2_b'], name='biases', trainable= self.trainable)
            out = tf.nn.bias_add(conv, biases)
            self.conv4_2 = tf.nn.relu(out)
            self.parameters += [kernel, biases]

        # conv4_3
        with tf.variable_scope('conv4_3',reuse=tf.AUTO_REUSE) as scope:
            kernel = tf.get_variable(initializer = self.weights['conv4_3_W'], name='weights', trainable= self.trainable)
            conv = tf.nn.conv2d(self.conv4_2, kernel, [1, 1, 1, 1], padding='SAME')
            biases = tf.get_variable(initializer = self.weights['conv4_3_b'], name='biases', trainable= self.trainable)
            out = tf.nn.bias_add(conv, biases)
            self.conv4_3 = tf.nn.relu(out)
            self.parameters += [kernel, biases]

        # pool4
        self.pool4 = tf.nn.max_pool(self.conv4_3,
                               ksize=[1, 2, 2, 1],
                               strides=[1, 2, 2, 1],
                               padding='SAME',
                               name='pool4')

        # conv5_1
        with tf.variable_scope('conv5_1',reuse=tf.AUTO_REUSE) as scope:
            kernel = tf.get_variable(initializer = self.weights['conv5_1_W'], name='weights', trainable= self.trainable)
            conv = tf.nn.conv2d(self.pool4, kernel, [1, 1, 1, 1], padding='SAME')
            biases = tf.get_variable(initializer = self.weights['conv5_1_b'], name='biases', trainable= self.trainable)
            out = tf.nn.bias_add(conv, biases)
            self.conv5_1 = tf.nn.relu(out)
            self.parameters += [kernel, biases]

        # conv5_2
        with tf.variable_scope('conv5_2',reuse=tf.AUTO_REUSE) as scope:
            kernel = tf.get_variable(initializer = self.weights['conv5_2_W'], name='weights', trainable= self.trainable)
            conv = tf.nn.conv2d(self.conv5_1, kernel, [1, 1, 1, 1], padding='SAME')
            biases = tf.get_variable(initializer = self.weights['conv5_2_b'], name='biases', trainable= self.trainable)
            out = tf.nn.bias_add(conv, biases)
            self.conv5_2 = tf.nn.relu(out)
            self.parameters += [kernel, biases]

        # conv5_3
        with tf.variable_scope('conv5_3',reuse=tf.AUTO_REUSE) as scope:
            kernel = tf.get_variable(initializer = self.weights['conv5_3_W'], name='weights', trainable= self.trainable)
            conv = tf.nn.conv2d(self.conv5_2, kernel, [1, 1, 1, 1], padding='SAME')
            biases = tf.get_variable(initializer = self.weights['conv5_3_b'], name='biases', trainable= self.trainable)
            out = tf.nn.bias_add(conv, biases)
            self.conv5_3 = tf.nn.relu(out)
            self.parameters += [kernel, biases]

        # pool5
        self.pool5 = tf.nn.max_pool(self.conv5_3,
                               ksize=[1, 2, 2, 1],
                               strides=[1, 2, 2, 1],
                               padding='SAME',
                               name='pool4')

    def fc_layers(self):
        # fc1
        with tf.variable_scope('fc1',reuse=tf.AUTO_REUSE) as scope:
            shape = int(np.prod(self.pool5.get_shape()[1:]))
            fc1w = tf.get_variable(initializer = self.weights['fc6_W'], name='weights')
            fc1b = tf.get_variable(initializer = self.weights['fc6_b'], name='biases')
            pool5_flat = tf.reshape(self.pool5, [-1, shape])
            fc1l = tf.nn.bias_add(tf.matmul(pool5_flat, fc1w), fc1b)
            self.fc1 = tf.nn.relu(fc1l)
            self.parameters += [fc1w, fc1b]
        
        # fc2
        with tf.variable_scope('fc2',reuse=tf.AUTO_REUSE) as scope:
            fc2w = tf.get_variable(initializer = self.weights['fc7_W'], name='weights')
            fc2b = tf.get_variable(initializer = self.weights['fc7_b'], name='biases')
            fc2l = tf.nn.bias_add(tf.matmul(self.fc1, fc2w), fc2b)
            self.fc2 = tf.nn.relu(fc2l)
            self.parameters += [fc2w, fc2b]

        # fc3
        with tf.variable_scope('fc3',reuse=tf.AUTO_REUSE) as scope:
            fc3w = tf.get_variable(initializer = self.weights['fc8_W'], name='weights')
            fc3b = tf.get_variable(initializer = self.weights['fc8_b'], name='biases')
            self.fc3l = tf.nn.bias_add(tf.matmul(self.fc2, fc3w), fc3b)
            self.parameters += [fc3w, fc3b]

    def load_weights(self, weight_file, sess):
        weights = np.load(weight_file)
        keys = sorted(weights.keys())
        for i, k in enumerate(keys):
            print(i, k, np.shape(weights[k]))
            sess.run(self.parameters[i].assign(weights[k]))
```

```
from tqdm import tqdm_notebook
def occlusion(output_layer,org_img,patch=50):
    with tf.Session() as sess:
        sess.run(tf.global_variables_initializer())
        cls_  = sess.run([output_layer], feed_dict={vgg.imgs:org_img.astype(np.float64).reshape([-1,224,224,3])})
        cls_ = np.array(cls_).squeeze()
        class_unit = output_layer[:,cls_.argmax()]
        mask = np.array([[[0.0]*3]*patch]*patch)
        mask = np.array(mask,np.float64)
        prob_dist = np.zeros((224,224,3))
        divide = np.zeros((224,224,3))
        temp = np.zeros((patch,patch,3))+1
        for i in tqdm_notebook(range(224-patch+1),desc='row position: '):
            for j in range(224-patch+1):
                # mask 10x10 patch
                divide[i:i+patch,j:j+patch,:]+=temp
                img = np.array(org_img).astype(np.float64)
                img[i:i+patch,j:j+patch,:]*=mask
                prob = sess.run([class_unit], feed_dict={vgg.imgs:img.reshape([-1,224,224,3])})
                prob_dist[i:i+patch,j:j+patch,:]+=(mask+prob[0])
        print("\nThe image class is : '"+class_names[cls_.argmax()])
        return prob_dist/divide
```

##5c. Guided backprop

```
from tensorflow.python.framework import ops
from tensorflow.python.ops import gen_nn_ops

@tf.RegisterGradient("GuidedRelu")
def _GuidedReluGrad(op, grad):
    gate_g = tf.cast(grad > 0, "float32")
    gate_y = tf.cast(op.outputs[0] > 0, "float32")
    return grad * gate_g * gate_y
```

```
def Guided_backprop(org_img,sess,output_layer,iters=1):
    sess.run(tf.global_variables_initializer())
    cls_  = sess.run([output_layer], feed_dict={vgg.imgs:org_img.astype(np.float64).reshape([-1,224,224,3])})
    cls_ = np.array(cls_).squeeze()
    gradient = tf.gradients(output_layer[:,cls_.argmax()],vgg.imgs)
    grad_mtx = np.zeros((224,224,3))
    org_img = org_img.astype(np.float64)
    for i in tqdm(range(iters)):
        grad = sess.run([gradient], feed_dict={vgg.imgs:org_img.reshape([-1,224,224,3])})
        grad = np.array(grad).squeeze()
        grad[grad<0] = 0.0
        grad_mtx+=grad
        org_img+=grad
        org_img = np.clip(org_img,0,255)
    print("\nThe image class is : '"+class_names[cls_.argmax()])
    return grad_mtx

tf.reset_default_graph()
with tf.Session() as sess:   
    g = tf.get_default_graph()
    with g.gradient_override_map({'Relu': 'GuidedRelu'}):
        imgs = tf.placeholder(tf.float32, [None, 224, 224, 3])
        vgg = vgg16(imgs, dep_path+'vgg16_weights.npz', sess)
        output_layer = vgg.fc3l
        backprop_guided = Guided_backprop(org_img,sess,output_layer)
        
```

##6. Grad-CAM

```
conv = vgg.conv5_3
output_layer = vgg.fc3l
```

```
# ref: https://deepvision-tensorflow.readthedocs.io/en/latest/_modules/tensorcv/models/layers.html
def global_avg_pool(x, name='global_avg_pool', data_format='NHWC'):
    assert x.shape.ndims == 4
    assert data_format in ['NHWC', 'NCHW']
    with tf.name_scope(name):
        axis = [1, 2] if data_format == 'NHWC' else [2, 3]
        return tf.reduce_mean(x, axis)

def Grad_cam(conv,output_layer,org_img,sess):
    sess.run(tf.global_variables_initializer())
    cls_  = sess.run([output_layer], feed_dict={vgg.imgs:org_img.astype(np.float64).reshape([-1,224,224,3])})
    cls_ = np.array(cls_).squeeze()
    #print(cls_.argsort()[-10:])
    feature_grad = tf.gradients(output_layer[:,cls_.argmax()],conv)
    feature_w = tf.convert_to_tensor(feature_grad)
    feature_w = global_avg_pool(tf.squeeze(feature_w,axis=0))
    conv_h = tf.shape(conv)[1]
    conv_w = tf.shape(conv)[2]
    conv_c = tf.shape(conv)[-1]
    conv_reshape = tf.reshape(conv,[conv_h*conv_w,conv_c])
    feature_w_reshape = tf.reshape(feature_w, [conv_c,1])
    class_map = tf.matmul(conv_reshape,feature_w_reshape)
    class_map = tf.reshape(class_map,[-1, conv_h,conv_w, 1])
    class_map = tf.nn.relu(tf.image.resize_bilinear(class_map, [224, 224]))
    classmap_out = sess.run([class_map], feed_dict={vgg.imgs:org_img.astype(np.float64).reshape([-1,224,224,3])})
    return np.array(classmap_out).squeeze()

with tf.Session() as sess:
    class_map = Grad_cam(conv,output_layer,org_img,sess)
```
